```python
# For Actors
for i in range(0,3146):
    temp=str(web_series.iloc[i]['actors']).split("Stars: ")
    try:
        web_series.iloc[i]['actors']=temp[1]
    except IndexError:
        web_series.iloc[i]['actors']='NA'
```

```python
# Splitting Genre
for i in range(0,3146):
    temp=str(web_series.iloc[i]['genre']).replace(",","")
    l=temp.split()
    try:
        web_series.iloc[i]['genre']=l
    except Exception as e:
        web_series.iloc[i]['genre']='NA'
```

```python
# Splitting Actors
for i in range(0,3146):
    temp=web_series.iloc[i]['actors']
    l=temp.split(",")
    try:
        web_series.iloc[i]['actors']=l
    except Exception as e:
        web_series.iloc[i]['actors']='NA'
```

```python
# Fixing url
for i in range(0,3146):
    web_series.iloc[i]['url']=web_series.iloc[i]['url'].replace("'","").replace("[","").replace("]","")
```

```python
# Exporting web series
import pickle
pickle.dump(web_series.to_dict(), open('web_series.pkl','wb'))
```

```python
# Deleting Space
for i in range(0,3146):  
    for j in range(0,len(web_series.iloc[i]['actors'])):
        web_series.iloc[i]['actors'][j]=web_series.iloc[i]['actors'][j].replace(" ","")
```

```python
web_series.head(4)
```

```python
# Splitting Web Series's Descriptions
web_series['description']=web_series['description'].apply(lambda x:x.split())
```

```python
# Creating new column 'tags'
web_series['tags'] =web_series['genre']  + web_series['actors'] + web_series['description']
```

```python
# Dropping none required columns
web_series=web_series.drop(columns=['genre','description','actors'])
```

```python
# Combining tags
web_series['tags'] = web_series['tags'].apply(lambda x:" ".join(x))
```

```python
# Making tags lowercase
web_series['tags']= web_series['tags'].apply(lambda x:x.lower())
```

```python
web_series['tags'].head(4)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemming(text):
    List=[]
    for i in text.split():
        List.append(ps.stem(i))
    return " ".join(List)
```

```python
web_series['tags']= web_series['tags'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3200, stop_words='english')
vectors=cv.fit_transform(web_series['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(web_show):
    index=web_series[web_series['title']==web_show].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distance[1:10]:
        print(web_series.iloc[i[0]].title)
```

```python
recommend("My Mister")
```
