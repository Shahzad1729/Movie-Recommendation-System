```python
movies['actors']=movies['crew']
movies['director']=movies['crew']
```

```python
movies=movies.drop(columns=['crew'])
```

```python
movies.head(5)
```

```python
# For Actors
for i in range(0,2911):
    temp=str(movies.iloc[i]['actors']).split("|")
    try:
        movies.iloc[i]['actors']=temp[1].replace("Stars: ","")
    except IndexError:
        movies.iloc[i]['actors']='NA'
```

```python
# For Director
for i in range(0,2911):
    temp=str(movies.iloc[i]['director']).split("|")
    try:
        movies.iloc[i]['director']=temp[0].replace("Director: ","")
    except IndexError:
        movies.iloc[i]['director']='NA'
```

```python
# Splitting Genre
for i in range(0,2911):
    temp=str(movies.iloc[i]['genre']).replace(",","")
    l=temp.split()
    try:
        movies.iloc[i]['genre']=l
    except Exception as e:
        movies.iloc[i]['genre']='NA'
```

```python
# Splitting Actors
for i in range(0,2911):
    temp=movies.iloc[i]['actors']
    l=temp.split(",")
    try:
        movies.iloc[i]['actors']=l
    except Exception as e:
        movies.iloc[i]['actors']='NA'
```

```python
# Splitting Director
for i in range(0,2911):
    l=[]
    temp=movies.iloc[i]['director']
    l.append(temp)
    try:
        movies.iloc[i]['director']=l
    except Exception as e:
        movies.iloc[i]['director']='NA'
```

```python
# Fixing url
for i in range(0,2911):
    movies.iloc[i]['url']=movies.iloc[i]['url'].replace("'","").replace("[","").replace("]","")
```

```python
# Exporting movies
import pickle
pickle.dump(movies.to_dict(), open('movies.pkl','wb'))
```

```python
# Deleting Space
for i in range(0,2911):  
    for j in range(0,len(movies.iloc[i]['actors'])):
        movies.iloc[i]['actors'][j]=movies.iloc[i]['actors'][j].replace(" ","")
```

```python
for i in range(0,2911):  
    for j in range(0,len(movies.iloc[i]['director'])):
        movies.iloc[i]['director'][j]=movies.iloc[i]['director'][j].replace(" ","")
```

```python
# Splitting Movie's Descriptions
movies['description']=movies['description'].apply(lambda x:x.split())
```

```python
# Creating new column 'tags'
movies['tags'] =movies['genre']  + movies['actors'] + movies['director']+movies['description']
```

```python
# Dropping none required columns
movies=movies.drop(columns=['genre','description','actors','director'])
```

```python
# Combining tags
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
# Making tags lowercase
movies['tags']= movies['tags'].apply(lambda x:x.lower())
```

```python
movies['tags']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemming(text):
    List=[]
    for i in text.split():
        List.append(ps.stem(i))
    return " ".join(List)
```

```python
movies['tags']= movies['tags'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000, stop_words='english')
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    index=movies[movies['title']==movie].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distance[1:10]:
        print(movies.iloc[i[0]].title)
```

```python
recommend("K.G.F: Chapter 1")
```
